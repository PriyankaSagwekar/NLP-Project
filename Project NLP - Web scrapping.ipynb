## Analysing Customer Reviews using NLP approach

The web has made it possible for organizations to discover what customers are saying about their brands online, either in mainstream media like online newspapers and magazines, or on social media. 

Consumers now search for opinions online before, during, and after a purchase. The next step for brands is finding out whether people are talking positively or negatively about their brand, and why. 

Some online ratings provide a number but not the reasoning behind it, and may only present half of the story. The process of analyzing a user's opinion or sentiment about particular services or products and their features is called opinion mining or sentiment analysis. 

It became an important task for online merchants to enable their customers to review or to express opinions on the products that they have purchased and use this feature to enhance customer experience.

### Web scrapping

```python
import pandas as pd
import requests # to send url request to website server
from bs4 import BeautifulSoup as bs # used to parse the html response from server into a suitable format 
```

```python
review = [] # Create an empty list 
```

```python
# save website url
url = "https://www.amazon.in/Inalsa-Fryer-Light-1400-Watt-Technology/product-reviews/B07GXQQP2Z/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews" 

# A function which accepts url and returns soup object containing parsed html response (a structured format of html code and web contents)
def get_soup(url):
    response = requests.get('http://localhost:8050/render.html',params={'url':url,'wait':2})
    soup = bs(response.content,'html.parser')
    return soup
```

```python
# title of html webpage
soup.title.text
```

```python
# A function that accepts soup object and returns all review details on that webpage
def get_reviews(soup):
    reviews = soup.find_all('div',{'data-hook':'review'})
    for item in reviews:
        r = {
        'title': item.find('a',{'data-hook':'review-title'}).text.strip(),
        'rating': float(item.find('i',{'data-hook':'review-star-rating'}).text.replace(' out of 5 stars','').strip()),
        'body': item.find('span',{'data-hook':'review-body'}).text.strip(),
         #'helped': int(item.find('span',{'data-hook':'helpful-vote-statement'}).text.replace(' people found this helpful','').strip()),
        'date': item.find('span',{'data-hook':'review-date'}).text.replace('Reviewed in India on ','').strip()
        }
        review.append(r)
```

```python
# call the function and scrape first 10 reviews on that webpage
soup = get_soup(url)
get_reviews(soup)
print(len(review))
```

```python
print(review)
```

```python
# a loop to iterate over all review pages available, scrape reviews on each page and append the review list

for i in range(2,60):
    soup = get_soup(f'https://www.amazon.in/Inalsa-Fryer-Light-1400-Watt-Technology/product-reviews/B07GXQQP2Z/ref=cm_cr_getr_d_paging_btm_next_{i}?ie=UTF8&reviewerType=all_reviews&pageNumber={i}')
    get_reviews(soup)
    print(f'Getting page: {i} \n {len(review)}')
    if not soup.find('li',{'class':'a-disabled a-last'}): #Last page has this particular tag which when encountered we break the loop
        pass
    else:
        break
```

```python
# converting our review list with necessary information into a dataframe df
import pandas as pd
df = pd.DataFrame(review)
df.to_csv('Product_reviews.csv',index=False)
```
